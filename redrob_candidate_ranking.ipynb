```python
import json

path = "challenge_data/[PUB] India_runs_data_and_ai_challenge/India_runs_data_and_ai_challenge/candidates.jsonl"

def calculate_score(candidate):

    score = 0

    # Experience Score (20)
    exp = candidate["profile"].get("years_of_experience", 0)

    if 5 <= exp <= 9:
        score += 20
    elif 4 <= exp < 5 or 9 < exp <= 11:
        score += 10

    # Skills Score (40)
    skills = [s["name"].lower() for s in candidate.get("skills", [])]

    important_skills = [
        "nlp",
        "fine-tuning llms",
        "milvus",
        "pinecone",
        "weaviate",
        "qdrant",
        "faiss",
        "embeddings",
        "retrieval",
        "ranking"
    ]

    for skill in important_skills:
        if skill in skills:
            score += 4

    # Open to Work (10)
    if candidate["redrob_signals"].get("open_to_work_flag"):
        score += 10

    # Recruiter Response Rate (10)
    score += candidate["redrob_signals"].get("recruiter_response_rate", 0) * 10

    # Github Activity (10)
    score += min(
        candidate["redrob_signals"].get("github_activity_score", 0),
        10
    )

    return round(score, 2)


top_candidates = []

with open(path, "r") as f:
    for line in f:
        candidate = json.loads(line)

        score = calculate_score(candidate)

        top_candidates.append(
            (
                candidate["candidate_id"],
                score,
                candidate["profile"]["current_title"]
            )
        )

top_candidates = sorted(
    top_candidates,
    key=lambda x: x[1],
    reverse=True
)

print("Top 20 Candidates")
for row in top_candidates[:20]:
    print(row)
```

```python
def calculate_score(candidate):

    score = 0

    # Experience
    exp = candidate["profile"].get("years_of_experience", 0)

    if 5 <= exp <= 9:
        score += 20
    elif 4 <= exp < 5 or 9 < exp <= 11:
        score += 10

    # Skills
    skills = [s["name"].lower() for s in candidate.get("skills", [])]

    important_skills = [
        "nlp",
        "fine-tuning llms",
        "milvus",
        "pinecone",
        "weaviate",
        "qdrant",
        "faiss"
    ]

    for skill in important_skills:
        if skill in skills:
            score += 5

    # Title Bonus/Penalty
    title = candidate["profile"].get("current_title", "").lower()

    good_titles = [
        "ai",
        "ml",
        "machine learning",
        "recommendation",
        "retrieval",
        "search",
        "ranking",
        "nlp",
        "applied"
    ]

    if any(word in title for word in good_titles):
        score += 20
    else:
        score -= 20

    # Open to work
    if candidate["redrob_signals"].get("open_to_work_flag"):
        score += 10

    # Response rate
    score += candidate["redrob_signals"].get(
        "recruiter_response_rate", 0
    ) * 10

    # Github
    score += min(
        candidate["redrob_signals"].get(
            "github_activity_score", 0
        ),
        10
    )

    return round(score, 2)
```

```python
import pandas as pd
import json

results = []

with open(path, "r") as f:
    for line in f:
        candidate = json.loads(line)

        score = calculate_score(candidate)

        results.append({
            "candidate_id": candidate["candidate_id"],
            "score": score,
            "title": candidate["profile"]["current_title"],
            "exp": candidate["profile"]["years_of_experience"]
        })

df = pd.DataFrame(results)

df = df.sort_values("score", ascending=False).reset_index(drop=True)

df["rank"] = range(1, len(df) + 1)

df.head(20)
```
